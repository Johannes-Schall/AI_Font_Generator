```python
# Now that DSRFontsDB is up to date, delete double entries
def load_json(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

def delete_files(json_data):
    for full_path in json_data:
        if os.path.exists(full_path):
            try:
                os.remove(full_path)
                print(f"Deleted: {full_path}")
            except Exception as e:
                print(f"Error deleting {full_path}: {e}")
        else:
            print(f"Could not find: {full_path}")

path_data = load_json(os.path.join(g.PATH_RAW, 'double_files.json'))
delete_files(path_data)
path_data = load_json(os.path.join(g.PATH_RAW, 'double_files2.json'))
delete_files(path_data)
```

```python
def normalize_path(path):
    return os.path.normpath(path)


# Load and apply clip filter
with open(os.path.join(g.PATH_RAW, 'clip.json'), 'r') as f:
    clip_data = json.load(f)
    
with open(g.PATH_TO_JSON_FONT_DB, 'r') as f:
    db_data = json.load(f)

clip_data_normalized = {normalize_path(key): value for key, value in clip_data.items()}
database_data_normalized = {normalize_path(key): value for key, value in db_data.items()}

    
for key, value in clip_data_normalized.items():
    if key in database_data_normalized:
        database_data_normalized[key].update(value)

# Write db
with open(g.PATH_TO_JSON_FONT_DB, 'w') as file:
    json.dump(database_data_normalized, file, indent=4) 
```
